```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("/Users/jaypatel/Desktop/hackathon/36-Hour-Hackathon-KIF-x-UIT/subscription_with_recency.csv")
```

```python
# Convert date
df["date"] = pd.to_datetime(df["date"])
```

```python
total_spending = df["amount"].sum()
category_spending = df.groupby("category")["amount"].sum()
merchant_spending = df.groupby("merchant")["amount"].sum().sort_values(ascending=False)
```

```python
# Frequency (number of payments per subscription)
frequency = df.groupby(["user_id", "merchant"]).size().reset_index(name="frequency")

# Recency (already exists, take latest per subscription)
recency = df.groupby(["user_id", "merchant"])["recency_days"].min().reset_index()

# Merge frequency + recency
metrics = pd.merge(frequency, recency, on=["user_id", "merchant"])
```

```python
#logic

def get_status(row):
    if row["recency_days"] > 40:
        return "INACTIVE"
    elif row["frequency"] >= 3:
        return "ACTIVE"
    else:
        return "UNCERTAIN"

metrics["status"] = metrics.apply(get_status, axis=1)
```

```python
# high cost 
high_cost = merchant_spending[merchant_spending > 1000]
```

```python
#duplicates
duplicates = df.groupby(["household_id", "merchant"]).size()
duplicates = duplicates[duplicates > 1]
```

```python

print("Total Spending:", total_spending)

print("\nCategory Spending:\n", category_spending)

print("\nTop Subscriptions:\n", merchant_spending.head())

print("\nHigh Cost Subscriptions:\n", high_cost)

print("\nDuplicates (Household):\n", duplicates)

print("\nSubscription Status:\n", metrics.head(10))
```
