```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#dataset loading
df = pd.read_csv("/Users/jaypatel/Desktop/hackathon/36-Hour-Hackathon-KIF-x-UIT/subscription_with_recency.csv")
```

```python
#Piechart
category_spending = df.groupby("category")["amount"].sum()
```

```python
plt.figure()
category_spending.plot(kind='pie', autopct='%1.1f%%')
plt.title("Spending by Category")
plt.ylabel("")
plt.show()
```

```python
#subscription bar
merchant_spending = df.groupby("merchant")["amount"].sum().sort_values(ascending=False)

plt.figure()
merchant_spending.head(10).plot(kind='bar')
plt.title("Top 10 Subscriptions by Spending")
plt.xlabel("Service")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
#recency plot
plt.figure()
df["recency_days"].plot(kind='hist', bins=10)
plt.title("Recency Distribution (Inactivity Indicator)")
plt.xlabel("Days Since Last Payment")
plt.ylabel("Count")
plt.show()
```
